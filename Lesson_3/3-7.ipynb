```r
library(GLMsData)
data(nhospital)
```

```r
str(nhospital)
```

```r
scatter.smooth(nhospital$MainHours ~ nhospital$Cases)
```

```r
scatter.smooth(nhospital$MainHours ~ nhospital$Eligible)
```

```r
scatter.smooth(nhospital$MainHours ~ nhospital$OpRooms)
```

```r
lm.1 <- lm(MainHours ~ Cases + Eligible, data=nhospital)
```

```r
summary(lm.1)
```

```r
summary(lm.1)$sigma
```

```r
lm.2 <- lm(MainHours ~ Cases + OpRooms, data=nhospital)
```

```r
summary(lm.2)
```

```r
sqrt(sum(resid(lm.2)^2)/df.residual(lm.2))
```

```r
summary(lm.2)$sigma
```

```r
lm.3 <- lm(MainHours ~ Eligible + OpRooms, data=nhospital)
```

```r
summary(lm.3)$sigma
```

```r
mu.1 <- fitted(lm.1)
mu.2 <- fitted(lm.2)
mu.3 <- fitted(lm.3)
```
